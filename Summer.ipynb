# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("summer.csv")
```

## 1. In how many cities Summer Olympics is held so far?

```python
print("In" , len(df['City'].unique()) ,"cities Summer Olympics is held so far")
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df[df.Medal == 'Gold'].groupby('Sport').Medal.count().sort_values(ascending = False).head().plot(x = 'Sport', y = 'medal', kind = 'bar', figsize = (20,5))
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python

data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])
    
pd.DataFrame(data,columns = ['Sport','medal']).sort_values(by='medal', ascending=False).head().plot(x = 'Sport', y = 'medal', kind = 'bar', figsize = (20,5))
```

## 4. Which player has won most number of medals? (Top 5) 

```python

data = []

for Medal in df['Athlete'].unique():
    data.append([Medal , len(df[df['Athlete']  == Medal])])
    
data = pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (20,5))
```

## 5. Which player has won most number Gold Medals of medals? (Top 5) 

```python
df[df.Medal == 'Gold'].groupby('Athlete').Medal.count().sort_values(ascending = False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (20,5))
```

## 6. In which year India won first Gold Medal in Summer Olympics? 

```python
year = df[(df.Medal == 'Gold') & (df.Country == 'IND')]['Year'].head().unique()
print(" In ", year[0] ,"India won first Gold Medal in Summer Olympics.")
```

## 7. Which event is most popular in terms on number of players? (Top 5) 

```python
data = []
for event in df['Event'].unique():
    data.append([event , len(df[df['Event'] == event])])
    
data = pd.DataFrame(data,columns = ['Event','Athlete']).sort_values(by = 'Athlete',ascending = False).head().plot(x = 'Event' , y = 'Athlete',kind = 'bar',figsize = (20,5))
```

## 8. Which sport is having most female Gold Medalists? (Top 5) 

```python
df[(df.Medal == 'Gold') & (df.Gender == 'Women')].groupby('Sport').Medal.count().sort_values(ascending = False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (20,5))
```
